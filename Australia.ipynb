```python
df['Total'] = df.sum(axis =1)
```

```python
df.describe()
```

```python
df.set_index('Country', inplace=True)
df.head()
```

```python
df.columns = list(map(str, df.columns))
 
for x in df.columns.values:
    print (type(x))
```

```python
years = list(map(str, range(1980, 2009)))
years
```

```python
df.sort_values(by = 'Total', ascending = False, axis = 0, inplace = True)
df_top3 = df.head(3)
df_top3 = df_top3[years].transpose()
df_top3
```

```python
df_top3.plot(kind='line', figsize=(10, 8)) 

plt.title('Immigration Trend of Top 5 Countries to Australia')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_top3.plot(kind='area', alpha = 0.5, stacked=False,figsize=(20, 10))  

plt.title('Immigration Trend of Top 5 Countries to Australia')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
