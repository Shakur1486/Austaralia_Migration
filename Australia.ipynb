```python
#set index of datafrane to Country
df.set_index('Country', inplace=True)
df.head()
```

```python
#convert columns to type string
df.columns = list(map(str, df.columns))

#print type of columns
for x in df.columns.values:
    print (type(x))
```

```python
#create list of years
years = list(map(str, range(1980, 2009)))
years
```

```python
#top 3 countries that migrated to Australia

df.sort_values(by = 'Total', ascending = False, axis = 0, inplace = True)
df_top3 = df.head(3)
df_top3 = df_top3[years].transpose()
df_top3
```

```python
#line plot showing the top 3 countries that migrated to Australia
df_top3.plot(kind='line', figsize=(10, 8)) 

plt.title('Immigration Trend of Top 3 Countries to Australia')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

It can be observed from the plot above, that the top 3 migrants that moved to 
Australia between 1980 to 2005 are from UK, New Zealand and United States of America.
United kingdom leads the pack and it can be observed from the plot that the migration 
from the UK to Australia is on the rise from 1984 to 2003 with only a small decline
experienced between 1990 to 1993. This meteoric rise in the movement from the UK to Australia
during this perios was contributed to by various factors ranging from economic Opportunities. 
Australia experienced strong economic growth during this period, leading to increased job 
opportunities in various sectors. The Australian economy was thriving, attracting skilled 
workers from the UK who sought better career prospects and higher wages.

```python
#Area plot showing the top 3 countries that migrated to Australia
df_top3.plot(kind='area', alpha = 0.5, stacked=False,figsize=(20, 10))  

plt.title('Immigration Trend of Top 5 Countries to Australia')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
#creating a dataframe for migrants from Africa
df_afr = df[df['Continent']=='Africa']
df_afr
```

```python
#assignining the top countries from Africa to a dataframe
df_afr_top5 = df_afr.head(5)
df_afr_top5 = df_afr_top5[years].transpose()
df_afr_top5
```

```python
#line plot showing the top 5 countries that migrated to Australia from Africa
df_afr_top5.plot(kind='line', figsize=(10, 8)) 

plt.title('Immigration Trend of Top 5 African Countries to Australia')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

It can be observed from the plot above, that the top 5 migrants that moved from Africa to 
Australia between 1980 to 2005 are from South Africa, Egypt, Zimbabwe, Kenya and Mauritius.
South Africa has the highest number of migrants from Africa during this period. A meteoric rise 
can be seen from 1994 till 2008 with only a small decline noticed between 2000 to 2004.
This meteoric rise in the movement from South Africa to Australia was influenced by a 
combination of political, economic, social, and personal factors.
South Africa experienced significant political changes during this period, 
including the end of apartheid and the establishment of a multiracial democracy. 
While the political transition was positive, some individuals still faced uncertainty and instability, 
leading them to seek better opportunities elsewhere. 
Also, the end of apartheid brought about social changes in South Africa, including affirmative action 
policies that favored previously disadvantaged groups. Some individuals from minority backgrounds 
felt that their opportunities were limited, leading them to consider migrating to Australia.

```python
#number of unique continents in the data
df['Continent'].unique()
```

```python
# group countries by continents and apply sum() function 
df_continents = df.groupby('Continent', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(df.groupby('Continent', axis=0)))

df_continents.head()
```

```python
##bar chart showing Immigration to Australia by continents

colors_list = ['deeppink', 'green', 'cyan', 'powderblue', 'red', 'orange']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

df_continents['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('Immigration to Australia by Continent [1980 - 2008]', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_continents.index, loc='upper left') 

plt.show()
```
